```python
# simple linear regression - Linear Algorithm 
# importing the pandas library to utilise functions related to the database
import pandas as pd
```

```python
# read the csv file(Dataset to be used) using the read_csv function from the pandas library and assign it to the Dataset variable
Dataset=pd.read_csv("Salary_Data.csv")
Dataset
```

```python
# split input and output of the dataset
# mention the particular input field column name of dataset and assign it to independent variable  
independent=Dataset[["YearsExperience"]]
independent
```

```python
# mention the output field column name of the dataset and assign it to Dependent variable
Dependent=Dataset[["Salary"]]
Dependent
```

```python
# import the train_test_split function from model_selection class under sklearn library
# train_test_split function will return 4 values: input train, input test, output train, output test, so we need to assign them to 4 variables while calling train_test_split function
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,Dependent,test_size=0.30,random_state=0)
```

```python
# learning phase
# import linearregression class from linear_model of sklearn library
from sklearn.linear_model import LinearRegression
# assigning all the functions related to linear regression to regressor variable to access them later
regressor=LinearRegression()
# accessing fit funtion using dot operator and passing input variables as parameters
regressor.fit(X_train,Y_train)
```

```python
# to check the weight  calculated in the model by accessing coef_ function,return variable of the coef_ function will be assigned to weight
weight=regressor.coef_
weight
```

```python
# to check the bias value calculated in the model by accessing intercept_ function
bias=regressor.intercept_
bias
```

```python
# predict the output value by accessing/calling predict function by passing input values of test dataset
y_predict=regressor.predict(X_test)
```

```python
#import r2_score function from metrics class of the sklearn module
# evaluating the model 
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_predict)
r_score
```

```python
# if value of r_score is nearing 1 then the model is the best model, so proceeding to save the model
# pickle is used to save the model
import pickle
filename="finalized_model_linear.sav"
```

```python
# model is saved to the created file using dump function , here wb refers to write the model to the created file
pickle.dump(regressor, open(filename,'wb'))
```

```python
# loading the model and to check the prediction
# load function is used to load the saved model to the loaded_model
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))
#  predict the loaded model using predict function by passing input variable as parameter, here yearsexperience is the input variable i.e 15
result=loaded_model.predict([[12]])
```
