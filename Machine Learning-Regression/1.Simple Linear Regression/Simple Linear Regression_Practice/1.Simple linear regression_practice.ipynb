```python
import pandas as pd
```

```python
Dataset=pd.read_csv("Salary_Data.csv")
Dataset
```

```python
# input output split
independent=Dataset[["YearsExperience"]]
independent
```

```python
Dependent=Dataset[["Salary"]]
Dependent
```

```python
# test and train split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,Dependent,test_size=0.30,random_state=0)
```

```python
# learning phase
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# weight and bias value
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
y_predict=regressor.predict(X_test)
```

```python
# evaluation metrics
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, y_predict)
```

```python
r_score
```

```python
# saving the model
import pickle
file_name ="FinalizedModel_Linear.sav"
```

```python
pickle.dump(regressor,open("FinalizedModel_Linear.sav",'wb'))
```

```python
# loading the model to validate the result
loaded_model=pickle.load(open("FinalizedModel_Linear.sav",'rb'))
result=loaded_model.predict([[14]])
```
