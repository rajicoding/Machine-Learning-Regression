<h2 style="color: purple; font-family: Comic Sans MS;">Importing Libraries</h2>

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.model_selection import GridSearchCV
```

<h2 style="color: purple; font-family: Comic Sans MS;">Importing Dataset</h2>


```python
dataset=pd.read_csv("Preprocessed_dataset.csv")
dataset
```

<h2 style="color: purple; font-family: Comic Sans MS;">Feature selection using Kbest</h2>

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function for selectkbest</h3>

```python
def selectkbest(indep_X,dep_Y,n):
    test = SelectKBest(score_func=f_regression, k=n)
    fit1 = test.fit(indep_X,dep_Y)
    # select features using selectkbest for the dataset and return them
    #selectkfeatures=fit1.transform(indep_X)
    #return selectkfeatures
    indices=fit1.get_support(indices=True)
    return indices
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function for standardscalar</h3>


```python
def scalar(indep_X):
    sc=StandardScaler()
    indep_X = sc.fit_transform(indep_X)
    return indep_X
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function for r2_prediction using grid search</h3>

```python
def r2_prediction_grid(grid,indep_X,dep_Y):
    grid_predict=grid.predict(indep_X)
    from sklearn.metrics import r2_score
    r_score = r2_score(dep_Y,grid_predict)
    return r_score
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function for Linear regression using grid search</h3>

```python
def Linear_grid(indep_X, dep_Y):
    from sklearn.linear_model import LinearRegression
    param_grid={
    'fit_intercept': [True, False],
    'positive': [True, False]
    }
    grid=GridSearchCV(estimator=LinearRegression(), param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
    grid.fit(indep_X,dep_Y)
    r_score=r2_prediction_grid(grid,indep_X,dep_Y)
    return r_score
    
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function for SVM using grid search</h3>

```python
def SVM_grid(indep_X,dep_Y):
    from sklearn.svm import SVR
    param_grid={'kernel':['linear','rbf'],
                'C':[0.1,1,10,100,500],
                'max_iter':[50000],
                'epsilon':[0.1,0.5,1.0],
               'gamma':['scale']}
    grid= GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
    grid.fit(indep_X, dep_Y)
    r_score=r2_prediction_grid(grid,indep_X,dep_Y)
    return r_score
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function for Decision Tree using grid search</h3>

```python
def Decision_grid(indep_X, dep_Y):
    from sklearn.tree import DecisionTreeRegressor
    param_grid={'criterion':['squared_error','friedman_mse'],
                'max_features':[None,'sqrt','log2'],
                'max_depth':[4,6],
                'splitter':['best','random'],
                'random_state':[0,None]}
    grid= GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
    grid.fit(indep_X, dep_Y)
    r_score=r2_prediction_grid(grid,indep_X,dep_Y)
    return r_score
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function for Random Forest using grid search</h3>

```python
def Random_grid(indep_X, dep_Y):
    from sklearn.ensemble import RandomForestRegressor
    param_grid={'criterion':['squared_error','friedman_mse'],
                'max_features':[None,'sqrt','log2'],
                'max_depth':[4,5],
                'n_estimators':[100,50],
                'random_state':[0,None]}
    grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
    grid.fit(indep_X, dep_Y)
    r_score=r2_prediction_grid(grid,indep_X,dep_Y)
    return r_score
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Function to create table for the models</h3>

```python
def selectk_regression(acclin,accsvm,accdes,accrf): 
    dataframe=pd.DataFrame(index=['f_regression'],columns=['Linear_grid','SVM_grid','Decision_grid','Random_grid'])
    
    for number,idex in enumerate(dataframe.index):
        dataframe.loc[idex,'Linear_grid']=acclin[number]       
        dataframe.loc[idex,'SVM_grid']=accsvm[number]
        dataframe.loc[idex,'Decision_grid']=accdes[number]
        dataframe.loc[idex,'Random_grid']=accrf[number]
    return dataframe
    
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Defining Input and Output data</h3>

```python
df2=dataset.copy(deep=True)

indep_X=df2.drop('BodyFat', axis=1)
dep_Y=df2['BodyFat']
```

```python
kbest=selectkbest(indep_X,dep_Y,6)      

acclin=[]
accsvm=[]
accdes=[]
accrf=[]

```

<h3 style="color:darkblue; font-family: Comic Sans MS;">Main function to run models for selected k features</h3>

```python

X_scaled = scalar(indep_X)

for i in kbest:   
    X_sub = X_scaled[:, i].reshape(-1, 1)   
    r2_lin=Linear_grid(X_sub,dep_Y)
    acclin.append(r2_lin)
    r2_sl=SVM_grid(X_sub,dep_Y)    
    accsvm.append(r2_sl)
    r2_d=Decision_grid(X_sub,dep_Y)
    accdes.append(r2_d)
    r2_r=Random_grid(X_sub,dep_Y)
    accrf.append(r2_r)
    
    
```

<h3 style="color:darkblue; font-family: Comic Sans MS;">R_score of models when k=3</h3>

```python
# k=3
result=selectk_regression(acclin,accsvm,accdes,accrf)
result
```
