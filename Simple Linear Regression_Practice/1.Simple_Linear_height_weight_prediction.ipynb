```python
# import pandas library and matplotlib library for plotting graph
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Dataset=pd.read_csv("Height_Weight.csv")
Dataset
```

```python
independent=Dataset[["Height"]]
independent
```

```python
dependent=Dataset[["Weight"]]
dependent
```

```python
plt.scatter(independent,dependent)
plt.xlabel("Height",fontsize=20)
plt.ylabel("Weight",fontsize=20)
plt.show()
```

```python
# Test and train data split using train_test_split from model_selection of sklearn
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
# model creation
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# weight can be determined using coef_ function
weight=regressor.coef_
weight
```

```python
# bias
bias=regressor.intercept_
bias
```

```python
y_pred=regressor.predict(X_test)
```

```python
# Evaluation metrics for the model
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
r_score
```

```python
# As the r_score is nearing 1 we can save the model 
import pickle
file_name="Simple_Linear_height_weight.sav"
```

```python
# dump the model to the created file using dump function
pickle.dump(regressor,open("Simple_Linear_height_weight.sav","wb"))
```

```python
# load the model to predict the ouput 
loaded_model=pickle.load(open("Simple_Linear_height_weight.sav","rb"))
result=loaded_model.predict([[210]])
```
