## 10.Convert the normal distribution to the standard normal distribution for the body fat percentage column

```python
# Converting normal distribution to Standard normal distribution
# defining the function with function name stdNBgraph and pre_dataset to be passed as parameter
def stdNBgraph(dataset):
# import seaborn library for plotting the graph    
    import seaborn as sns
# calculating the mean for the given column in the dataset using mean() function and assign it to mean variable
    mean=dataset.mean()
# calculating the standard deviation for the given column in the dataset using std() function and assigne it to std variable
    std=dataset.std()
# each values of particular column in dataset is store in list i and assigned to  list values    
    values = [i for i in dataset]
# z_score is calculated for each values in dataset and stored to list z_score    
    z_score =[((j-mean)/std) for j in values]
# density plot with respect to the given column in the dataset is plotted using distplot method by passing z_score as parameters
# kde=True so the curve is drawn in the plot    
    sns.distplot(z_score,kde=True)
# summation of the z_score divided by number of z_score values is calculated    
    (sum(z_score)/len(z_score))
```

## 14. What is the probability Density Function of the body fat percentage range from 20 to 35%

```python
# function to display the probability density function for the given range
def get_pdf_probability(dataset,startrange,endrange):
#importing the required libraries for plotting the graph
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
# using seaborn plotting the graph with distplot method
# kde=True so the curve for probability density plot
# color for curve is given as red using kde _kws attribute
# color for the histogram plot is given as green using color attribute
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':"red"},color='Green')  
# vertical lines for startrange and endrange is drawn using axvline method of pyplot and color of the line is given as blue 
    pyplot.axvline(startrange,color='blue')
    pyplot.axvline(endrange,color='blue')
# generate a sample
    sample=dataset
# calculate the mean and standard deviation using  mean() and std() and assign it to sample_mean and sample_std variable    
    sample_mean=sample.mean()
    sample_std=sample.std()
# Formatting the mean and standard deviation upto 3 decimal and printing the same
    print('Mean=%.3f,Standard deviation=%.3f'%(sample_mean,sample_std))
# define the distribution
    dist=norm(sample_mean,sample_std)
# sample probabilities for a range of outcomes
# list named value is created, values of the list is generated between the given range(start and endrange by the user)  stored in the list named value and assigned back to list values    
    values=[value for value in range(startrange,endrange)] 
# adding print statement to compare with usual method of using for loop with 3 lines
   
# pdf method from dist class is used to calculate the probability for each value available in values list 
# and all the probability values are stored in probabilities list   
    probabilities=[float(dist.pdf(value)) for value in values]
#adding all the probabiltiy value of the given range and stored in to the variable prob  
    prob=(sum(probabilities))
#Printing the overall probability value from startrange to endrange provided by the user
    print("the area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    
```
